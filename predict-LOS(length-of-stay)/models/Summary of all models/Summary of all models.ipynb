```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import sklearn
```

```python
GB_models_untuned = pd.read_csv(r"../Gradient Boosting and Extreme Gradient Boosting/GB_models_untuned.csv")

DTR_RFR_models_untuned = pd.read_csv(r'../Decision Tree + Random Forest/DTR+RFR_models_untuned.csv')

linear_models_untuned = pd.read_csv(r'../Linear models/linear_models_untuned.csv')

SVR_KNR_models_untuned = pd.read_csv(r'../Support Vector + K Neighbors Regressor/SVR+KNR_models_untuned.csv')
```

```python
GB_models_tuned = pd.read_csv(r"../Gradient Boosting and Extreme Gradient Boosting/GB_models_tuned.csv")

DTR_RFR_models_tuned = pd.read_csv(r'../Decision Tree + Random Forest/DTR+RFR_models_tuned.csv')

linear_models_tuned = pd.read_csv(r'../Linear models/linear_models_tuned.csv')

SVR_KNR_models_tuned = pd.read_csv(r'../Support Vector + K Neighbors Regressor/SVR+KNR_models_tuned.csv')
```

```python
df_all_untuned_results = pd.merge(linear_models_untuned, DTR_RFR_models_untuned). merge(GB_models_untuned).merge(SVR_KNR_models_untuned)

df_all_untuned_results
```

```python
corr_matrix_all_untuned_results = df_all_untuned_results.drop(columns = ['PATIENT_NO']).corr()

plt.figure(figsize=(10,10))
fig = sns.heatmap(corr_matrix_all_untuned_results,
            vmin=-1,
            cmap='coolwarm',
            annot=True)
fig.figure.savefig('./corr_matrix_all_untuned_results.png', bbox_inches = 'tight')
```

```python
df_all_tuned_results = pd.merge(linear_models_tuned, DTR_RFR_models_tuned). merge(GB_models_tuned).merge(SVR_KNR_models_tuned)

df_all_tuned_results
```

```python
corr_matrix_all_tuned_results = df_all_tuned_results.drop(columns = ['PATIENT_NO']).corr()

plt.figure(figsize=(10,10))
fig = sns.heatmap(corr_matrix_all_tuned_results,
            vmin=-1,
            cmap='coolwarm',
            annot=True)
fig.figure.savefig('./corr_matrix_all_tuned_results.png', bbox_inches = 'tight')
```

```python
df_all_tuned_results.columns.values
```

```python
df_selected_results = df_all_tuned_results[['PATIENT_NO',
                                            'LOS_LinearRegression',
                                            'LOS_RandomForestRegressor',
                                            'LOS_GradientBoostingRegressor',
                                            'LOS_XGBRegressor',
                                            'LOS_SVRegression',]]

df_selected_results
```

```python
corr_matrix_selected_tuned_results = df_selected_results.drop(columns = ['PATIENT_NO']).corr()

plt.figure(figsize=(10,10))
fig = sns.heatmap(corr_matrix_selected_tuned_results,
            vmin=-1,
            cmap='coolwarm',
            annot=True)
fig.figure.savefig('./corr_matrix_selected_tuned_results.png', bbox_inches = 'tight')
```

```python
df_extreme_selected_results = df_all_tuned_results[['PATIENT_NO',
                                            'LOS_GradientBoostingRegressor',
                                            'LOS_XGBRegressor',
                                            'LOS_SVRegression',]]

df_extreme_selected_results
```

```python
df_extreme_selected_results.describe()
```

```python
df_extreme_selected_results['MIN_predicted_LOS'] = df_extreme_selected_results.loc[:, [
                           'LOS_GradientBoostingRegressor',
                           'LOS_XGBRegressor',
                           'LOS_SVRegression']].min(axis=1)


df_extreme_selected_results['MAX_predicted_LOS'] = df_extreme_selected_results.loc[:, [
                           'LOS_GradientBoostingRegressor',
                           'LOS_XGBRegressor',
                           'LOS_SVRegression']].max(axis=1)

df_extreme_selected_results['AVG_predicted_LOS'] = df_extreme_selected_results.loc[:, [
                           'LOS_GradientBoostingRegressor',
                           'LOS_XGBRegressor',
                           'LOS_SVRegression']].mean(axis = 1)

df_extreme_selected_results['diff_predicted_LOS'] =  df_extreme_selected_results['MAX_predicted_LOS'] - df_extreme_selected_results['MIN_predicted_LOS']
```
