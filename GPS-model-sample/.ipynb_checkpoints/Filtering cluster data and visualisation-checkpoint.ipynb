```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('./finished_data.csv')
```

```python
df = df.drop(df.columns[[0, 1]], axis=1)
df.head()
```

```python
df.describe()
```

```python
df['cluster'].head()
```

```python
df['device_id'].nunique()
```

```python
df_cluster_value_counts = df['cluster'].value_counts()
```

```python
df_clusters = df[df['cluster'].isin(df_cluster_value_counts[df_cluster_value_counts>1].index)]
```

```python
df_clusters.head()
```

```python
# test sample 
df_clusters[df_clusters['cluster'] == 13] 
```

```python
df_clusters.describe()
```

```python
df_clusters.to_csv('./finished_data_clusters.csv')
```

```python
df_gd = df.groupby('device_id')
```

```python
df_device_id_value_counts = df['device_id'].value_counts()
df_thres_5 = df[df['device_id'].isin(df_device_id_value_counts[df_device_id_value_counts >= 5].index)]
df_thres_5.describe()
```

```python
df_device_id_value_counts.head()
```

```python
df_thres_5['device_id'].nunique()
```

```python
# number of people
i = 11330
# number of threshold of points 
thresh_points = 5
```

```python
plt.figure(figsize=(25,20)) 

for g in df_gd:
    g = g[1]
    # Limit to china only
    g = g.loc[g.latitude > 18]
    g = g.loc[g.longitude > 80]
    if g.shape[0] > thresh_points:
        lon = g['longitude']
        lat = g['latitude']
        plt.plot(lon, lat, '-o', color=np.random.rand(3).tolist())
        i -= 1
    if i < 0:
        break
    
plt.savefig('./gps_map.png')
```

```python
df_clusters['device_id'].nunique()
```

```python
# number of people
i_clusters = df_clusters['device_id'].nunique()
# number of threshold of points 
thresh_points_clusters = 5
```

```python
plt.figure(figsize=(25,20)) 

for g in df_clusters.groupby('device_id'):
    g = g[1]
    if g.shape[0] > thresh_points_clusters:
        lon = g['longitude']
        lat = g['latitude']
        plt.scatter(lon, lat, 'o', c=df_clusters['cluster'].tolist())
        i_clusters -= 1
    if i_clusters < 0:
        break
    
plt.savefig('./gps_map_clusters.png')
```
