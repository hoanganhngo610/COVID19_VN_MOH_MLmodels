```python
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
```

```python
path = '/Users/thainam/Documents/GitHub/COVID19_VN_MOH_MLmodels/GPS-model-sample/events_250k.csv'
data = pd.read_csv(path)

data
# pd.set_option('display.float_format', lambda x: '%.0f' % x)
```

```python
merge = data.sort_values(by=['device_id', 'timestamp'])
merge
```

```python
timestamp = merge.iloc[:, 3:4].values
device_id = merge.iloc[:, 2:3].values
latitude = merge.iloc[:, 5:6].values #y
latitude = np.reshape(latitude, len(latitude)).tolist()
longitude = merge.iloc[:, 4:5].values #x
longitude = np.reshape(longitude, len(longitude))
```

```python
import math

arr = [0.0]
for i in range (1, len(latitude)):
    if device_id[i - 1] == device_id[i]:
        distance = math.sqrt(pow(longitude[i] - longitude[i - 1], 2) + pow(latitude[i] - latitude[i - 1], 2))
        arr.append(distance)
    elif device_id[i - 1] != device_id[i]:
        arr.append(0)
    else:
        continue
    
print(np.array(arr))
```

```python
arr = np.array(arr)
arr = np.reshape(arr, (len(arr), 1))
df = np.concatenate((device_id, np.array(arr)), axis=1)
df = pd.DataFrame(data=df, columns=['device_id', 'distance'])
df.head()
```

```python
distance = df.groupby(['device_id']).agg('sum')
distance
device_id = pd.DataFrame(data=device_id, columns=['device_id'])
device_id.drop_duplicates(subset ='device_id', 
                     keep = False, inplace = True)
device_id.sort_values(by=['device_id'])
# total = pd.concat([device_id, distance])
```
